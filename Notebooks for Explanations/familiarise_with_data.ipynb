#### __Categorical Variables__
It can be very useful to check if a variable is categorical or not. Converting variables to category data type is important for the visualisation and analysis part of this project, so whilst this process may be lengthy it gives a valuable outcome.

```python
# Check home_ownership values are categorical
print(loan_data['home_ownership'].nunique())
print(loan_data['home_ownership'].unique())
```

```python
# Check verification_status values are categorical
print(loan_data['verification_status'].nunique())
print(loan_data['verification_status'].unique())
```

```python
# Check loan_status values are categorical
print(loan_data['loan_status'].nunique())
print(loan_data['loan_status'].unique())
```

```python
print(loan_data['payment_plan'].nunique())
print(loan_data['payment_plan'].unique())
```

__payment_plan__ only has 2 unique values which indicate 'no' and 'yes', hence this variable can be converted to boolean datatype.

```python
print(loan_data['purpose'].nunique())
print(loan_data['purpose'].unique())
```

```python
print(loan_data['application_type'].nunique())
print(loan_data['application_type'].unique())
```

__application_type__ only has 1 unique value so instead of converting the variable to category datatype, it may be more useful to remove the variable from the data.

```python
print(loan_data['term'].nunique())
print(loan_data['term'].unique())
```

```python
print(loan_data['policy_code'].nunique())
print(loan_data['policy_code'].unique())
```

__policy_code__ only has 1 unique value, so again the variable does not bring much insight to our future analysis and so can be removed.
